 问题出现在尝试调用 tokenize 时。在 NLTK 中，tokenize 是一个模块，而不是一个可以直接调用的函数。    
 需要从 tokenize 模块中导入具体的分词函数，例如 word_tokenize 或 sent_tokenize。
